```python
import pandas as pd
import matplotlib.pyplot as plt

# загрузка csv
df = pd.read_csv("results.csv")

df.head()
```

```python
random_df = df[df["array_type"] == "random"]

plt.figure(figsize=(12, 7))

for algo in random_df["algorithm"].unique():

    subset = random_df[
        random_df["algorithm"] == algo
    ]

    plt.plot(
        subset["size"],
        subset["time_us"],
        label=algo
    )

plt.xlabel("Array size")
plt.ylabel("Time (microseconds)")
plt.title("Random arrays: execution time")
plt.legend()
plt.grid()

plt.show()
```

```python
reverse_df = df[df["array_type"] == "reverse"]

plt.figure(figsize=(12, 7))

for algo in reverse_df["algorithm"].unique():

    subset = reverse_df[
        reverse_df["algorithm"] == algo
    ]

    plt.plot(
        subset["size"],
        subset["time_us"],
        label=algo
    )

    plt.xlabel("Array size")
    plt.ylabel("Time (microseconds)")
    plt.title("Reverse arrays: execution time")
    plt.legend()
    plt.grid()

plt.show()
```

```python
almost_df = df[df["array_type"] == "almost_sorted"]

plt.figure(figsize=(12, 7))

for algo in almost_df["algorithm"].unique():

    subset = almost_df[
        almost_df["algorithm"] == algo
    ]

    plt.plot(
        subset["size"],
        subset["time_us"],
        label=algo
    )

plt.xlabel("Array size")
plt.ylabel("Time (microseconds)")
plt.title("Almost sorted arrays: execution time")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(12, 7))

for algo in df["algorithm"].unique():

    if algo == "MSDRadixSort":
        continue

    subset = df[
        (df["algorithm"] == algo) &
        (df["array_type"] == "reverse")
    ]

    plt.plot(
        subset["size"],
        subset["char_comparisons"],
        label=algo
    )

plt.xlabel("Array size")
plt.ylabel("Character comparisons")
plt.title("Reverse arrays: character comparisons")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(12, 7))

algorithms = [
    "QuickSort",
    "StringQuickSort"
]

for algo in algorithms:

    subset = df[
        (df["algorithm"] == algo) &
        (df["array_type"] == "reverse")
    ]

    plt.plot(
        subset["size"],
        subset["time_us"],
        linewidth=3,
        label=algo
    )

plt.xlabel("Array size")
plt.ylabel("Time (microseconds)")
plt.title("QuickSort vs StringQuickSort")
plt.legend()
plt.grid()

plt.show()
```
