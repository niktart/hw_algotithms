```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt

#hybrid_merge_sort
df_h_almost = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/hybrid_merge_almost.csv')
df_h_random = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/hybrid_merge_random.csv')
df_h_reverse = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/hybrid_merge_reverse.csv')

#merge_sort
df_m_almost = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/standard_merge_almost.csv')
df_m_random = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/standard_merge_random.csv')
df_m_reverse = pd.read_csv('/content/drive/MyDrive/aaaaa/analysis/A2/standard_merge_reverse.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_h_almost['Size'], df_h_almost['TimeMicroseconds'] / 1_000_000, label='Hybrid Merge Sort')
plt.plot(df_m_almost['Size'], df_m_almost['TimeMicroseconds'] / 1_000_000, label='Standard Merge Sort')

plt.title('Performance Comparison for "Almost Sorted" Data')
plt.xlabel('Input Size')
plt.ylabel('Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_h_random['Size'], df_h_random['TimeMicroseconds'] / 1_000_000, label='Hybrid Merge Sort')
plt.plot(df_m_random['Size'], df_m_random['TimeMicroseconds'] / 1_000_000, label='Standard Merge Sort')

plt.title('Performance Comparison for "Random" Data')
plt.xlabel('Input Size')
plt.ylabel('Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_h_reverse['Size'], df_h_reverse['TimeMicroseconds'] / 1_000_000, label='Hybrid Merge Sort')
plt.plot(df_m_reverse['Size'], df_m_reverse['TimeMicroseconds'] / 1_000_000, label='Standard Merge Sort')

plt.title('Performance Comparison for "Reverse Sorted" Data')
plt.xlabel('Input Size')
plt.ylabel('Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```

# Выводы
По графикам видно, что начиная с 20.000 и до 65.000 элементов в массиве гибридная сортировка работает лучше, чем обычный merge_sort

На почти отсортированном и рандомном массивах гибридный подход всегда выигрывает стандартный

На массиве, отсортированном в обратном порядке гибридный подход до 70.000 выигрывает, но потом ведет себя примерно также.
