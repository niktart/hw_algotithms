```python
#include <iostream>
#include <cmath>
#include <fstream>
#include <random>
#include <vector>

int main() {
    std::ios::sync_with_stdio(false);
    std::cin.tie(nullptr);

    std::vector<std::vector<double>> list(3, std::vector<double>(3));
    for (int i = 0; i < 3; i++)
    {
        for (int j = 0; j < 3; j++)
        {
            std::cin >> list[i][j];
        }
    }

    std::ofstream outfile("monte_carlo_data_small.csv");
    outfile << "N,calculated_area,deviation\n";

    // double x_min = INT_MAX, x_max = -INT_MAX;
    // double y_min = INT_MAX, y_max = -INT_MAX;
    //
    // //    
    // for (int i = 0; i < 3; i++) {
    //     x_min = std::min(x_min, list[i][0] - list[i][2]);
    //     x_max = std::max(x_max, list[i][0] + list[i][2]);
    //     y_min = std::min(y_min, list[i][1] - list[i][2]);
    //     y_max = std::max(y_max, list[i][1] + list[i][2]);
    // }

    double x_min = -INT_MAX, x_max = INT_MAX;
    double y_min = -INT_MAX, y_max = INT_MAX;

    //    
    for (int i = 0; i < 3; i++) {
        x_min = std::max(x_min, list[i][0] - list[i][2]);
        x_max = std::min(x_max, list[i][0] + list[i][2]);
        y_min = std::max(y_min, list[i][1] - list[i][2]);
        y_max = std::min(y_max, list[i][1] + list[i][2]);
    }
    double result = 0.44380799999999998029;

    for (int N = 100; N <= 100000; N += 500)
    {
        int inside = 0;

        std::random_device rd;
        std::mt19937 gen(rd());
        std::uniform_real_distribution<double> dx(x_min, x_max);
        std::uniform_real_distribution<double> dy(y_min, y_max);

        for (int i = 0; i < N; i++) {
            double gen_x = dx(gen);
            double gen_y = dy(gen);

            bool flag = true;
            for (int j = 0; j < 3; j++) {
                double x = list[j][0];
                double y = list[j][1];
                double r = list[j][2];

                double dx = gen_x - x;
                double dy = gen_y - y;
                if (dx * dx + dy * dy > r * r) {
                    flag = false;
                    break;
                }
            }
            if (flag)
            {
                inside++;
            }
        }

        double rectArea = (x_max - x_min) * (y_max - y_min);
        double area = static_cast<double>(inside) / N * rectArea;
        double deviation = std::abs(area - result);

        outfile << N << "," << area << "," << deviation << "\n";
    }

}
```
